```python
import scipy.stats as ss
def chi2(arr):
    R_N = arr.sum(axis = 1)
    C_N = arr.sum(axis = 0)
    N = arr.sum()
    E = np.ones(arr.shape) * C_N / N
    E = (E.T * R_N).T
    square = (arr - E) ** 2 / E
    square[E == 0] = 0
    v = square.sum()
    return v

def chi2Merge(df,col,target,max_groups=None,threshold=None):
    '''
    卡方分箱
    df: 数据集
    col: 需要分箱的变量名
    target: 类标签
    max_groups: 最大分组数目
    threshold: 卡方阈值，如果未指定max_groups,默认使用95%置信度设置阈值
    return: 包括各组的起始值列表
    '''
    freq_tab = pd.crosstab(df[col], df[target])
    freq = freq_tab.values
    cutoffs = freq_tab.index.values
    if max_groups is None:
        if threshold is None:
            cls_num = freq.shape[-1]
            threshold = ss.chi2.isf(0.05, df = cls_num - 1)
    while True:
        minvalue = None
        minidx = None
        for i in range(len(freq) - 1):
            v = chi2(freq[i : i+2])
            if minvalue is None or minvalue > v:
                minvalue = v
                minidx = i
        if (max_groups is not None and max_groups < len(freq)) or (threshold is not None and minvalue < threshold):
            tmp = freq[minidx] + freq[minidx + 1]
            freq[minidx] = tmp
            freq = np.delete(freq, minidx + 1, 0)
            cutoffs = np.delete(cutoffs, minidx + 1, 0)
        else:
            break
    return cutoffs

def value2group(x, cutoffs):
    '''
    将变量的值转换为响应的组
    x: 需要转换到分组的值
    cutoffs: 各组的起始值
    return: x对应的组，从group1开始
    '''
    cutoffs = sorted(cutoffs)
    num_groups = len(cutoffs)
    if x < cutoffs[0]:
        return 'group1'
    for i in range(1, num_groups):
        if cutoffs[i - 1] <= x < cutoffs[i]:
            return 'group{}'.format(i)
    return 'group{}'.format(num_groups)

```

```python
cutoffs = chi2Merge(df,'年龄','result',max_groups =10)
print(cutoffs)
df['age_group'] = [value2group(x,cutoffs) for x in df['年龄']]
df
```
