## 对比分析  
先处理表中的所有异常值，例如空值、离群值等等

```python
import pandas as pd
import numpy as np

df = pd.read_csv('~/mycode/python/data_analyze/data/HR.csv')
df.count()
```

```python
#删除空值
df = df.dropna(axis=0,how="any")#axis 0为行，1为列,how all全部为空，any只要任意一个为空
df.count()
```

```python
#删除异常值
df = df[df["last_evaluation"]<=1][df["salary"]!="nme"][df["department"]!="sale"]
df.count()
```

```python
df["satisfaction_level"].describe()
```

## 对比分析  
以部门为单位进行分组对比

```python
#按照部门分组，求各个组的均值，对于文字类属性，直接去掉
df.groupby("department").mean()
```

```python
#按照部门分组，求各个组的标准差
df.groupby("department").std()
```

```python
#切片分组
df.loc[:,["last_evaluation","department"]].groupby("department").mean()
```

```python
df.loc[:,["average_montly_hours","department"]].groupby("department")["average_montly_hours"].apply(lambda x:x.max()-x.min())
```
