## 可视化分析  
**python常见可视化工具** matplotlib,seaborn,plotly

```python
import numpy as np
import pandas as pd
import plotly as py
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot

init_notebook_mode(connected=True)

df = pd.read_csv('~/mycode/python/data_analyze/data/HR.csv')
df.head(10)
```

```python
np.arange(len(df["salary"].value_counts()))
#df["salary"].value_counts().index
df['salary'].value_counts()
```

```python
data = [go.Bar(x=df['salary'].value_counts().index,y=df['salary'].value_counts().values)]
iplot(data,filename='Salary')
```

```python
plt.title("salary")
plt.xlabel("salary")
plt.ylabel("number")
plt.xticks(np.arange(len(df["salary"].value_counts())), df["salary"].value_counts().index)
plt.bar(np.arange(len(df["salary"].value_counts())), df["salary"].value_counts(),width=0.5)
for x,y in zip(np.arange(len(df["salary"].value_counts())),df["salary"].value_counts()):
    plt.text(x,y,y,ha="center",va="bottom")
plt.show()
```


## PIE

```python
lbs = df["salary"].value_counts().index
explodes=[0.1 if i=="low" else 0 for i in lbs]
print(explodes)
plt.pie(df["salary"].value_counts(normalize=True),labels=lbs, autopct="%1.1f%%",explode=explodes)
plt.show()
```
