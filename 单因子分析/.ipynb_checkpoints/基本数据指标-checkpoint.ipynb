```python
import numpy as np
import pandas as pd

df = pd.read_csv('~/mycode/python/data_analyze/data/HR.csv')
df.head(10)
```

```python
#均值
df.mean()
```

```python
#中位数
df.median()
```

```python
#众数
df.mode()
```

```python
#分位数
df.quantile(q=0.25)
```

```python
#标准差
df.std()
```

```python
#方差
df.var()
```

```python
#求和
df.sum()
```

```python
#偏态系数
df.skew()
```

```python
#峰度系数
df.kurt()
```

```python
import scipy.stats as ss
```

```python
#生成正态分布
data = ss.norm()
data.stats(moments='mvsk')#均值、方差、偏态系数、峰度系数
```

```python
#pdf 给出横坐标的值，计算纵坐标的值，在正太分布中为0.39
ss.norm.pdf(0.0)
```

```python
#ppf 从负无穷大积分到1.28，值等于0.9
ss.norm.ppf(0.9)
```

```python
#cdf 从负无穷积分到给定值，计算概率
ss.norm.cdf(2)
```

```python
#计算正态分布冲-2积分到2的概率
ss.norm.cdf(2)-ss.norm.cdf(-2)
```

```python
#给出满足正态分布的数值
ss.norm.rvs(size=100)
```
