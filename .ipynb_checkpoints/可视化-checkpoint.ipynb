## 可视化分析  
**python常见可视化工具** matplotlib,seaborn,plotly

```python
import numpy as np
import pandas as pd

df = pd.read_csv('~/mycode/python/data_analyze/data/HR.csv')
df.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
np.arange(len(df["salary"].value_counts()))
#df["salary"].value_counts().index
```

```python
plt.title("salary")
plt.xlabel("salary")
plt.ylabel("number")
plt.xticks(np.arange(len(df["salary"].value_counts())), df["salary"].value_counts().index)
plt.bar(np.arange(len(df["salary"].value_counts())), df["salary"].value_counts(),width=0.5)
for x,y in zip(np.arange(len(df["salary"].value_counts())),df["salary"].value_counts()):
    plt.text(x,y,y,ha="center",va="bottom")
plt.show()
```

## seaborn

```python
plt.close()
sns.set_style(style="whitegrid")
plt.title("salary")
plt.xlabel("salary")
plt.ylabel("number")
plt.xticks(np.arange(len(df["salary"].value_counts())), df["salary"].value_counts().index)
plt.bar(np.arange(len(df["salary"].value_counts())), df["salary"].value_counts(),width=0.5)
for x,y in zip(np.arange(len(df["salary"].value_counts())),df["salary"].value_counts()):
    plt.text(x,y,y,ha="center",va="bottom")
plt.show()
```
